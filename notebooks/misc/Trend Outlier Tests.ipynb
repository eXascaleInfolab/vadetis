```python
import numpy as np
```

```python
factor = 8 / 10
timestamps = [(20, 150)]
for start, end in timestamps:
    additional_values = np.zeros(2*end)
    print("additional_values\n", additional_values)
    print(additional_values.size)
    slope = np.random.choice([-1, 1]) * factor * np.arange(end - start)
    print("slope\n", slope)
    print(slope.size)
    additional_values[list(range(0, 130))] += slope
    print("additional_values\n", additional_values)
    print("slope[-1]\n", slope[-1])
    print("additional_values[end:]", additional_values[end:])
    additional_values[end:] += slope[-1]
```
