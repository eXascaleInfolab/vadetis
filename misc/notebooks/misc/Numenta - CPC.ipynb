```python
import pandas as pd
import numpy as np
import json, csv, datetime
```

```python
#input

nab_path = '/home/adrian/Dokumente/real_data/NAB-master'
nab_data_path = nab_path + '/data/'
 
labels_filename = '/labels/combined_labels.json'
file_name_ex2 = 'realAdExchange/exchange-2_cpc_results.csv'
file_name_ex3 = 'realAdExchange/exchange-3_cpc_results.csv'
file_name_ex4 = 'realAdExchange/exchange-4_cpc_results.csv'

files = [
    #{'name': 'Exchange2', 'file_name': file_name_ex2},
    {'name': 'Exchange3', 'file_name': file_name_ex3},
    {'name': 'Exchange4', 'file_name': file_name_ex4},
]


#output
output_path = '/home/adrian/Dokumente/real_data/nab_cpc'
test_file_name = output_path + '/test_cpc.csv'
train_file_name = output_path + '/train_cpc.csv'
```

```python
labels_file = open(nab_path + labels_filename, 'r')
labels = json.loads(labels_file.read())
labels_file.close()

def load_data_frame_with_labels(ts_name, file_name):
    df = pd.read_csv(nab_data_path + file_name)
    df.insert(0, 'ts_name', ts_name)
    df.insert(2, 'unit', "Value")
    df['class'] = df['timestamp'].isin(labels[file_name]).astype(int)
    df = df.rename(columns={'timestamp': 'time'})
    return df
 
#training_data_frame = load_data_frame_with_labels(training_file_name)
#test_data_frame = load_data_frame_with_labels("Test", file_name_UPS)
```

```python
def init_file(outputfile):
    with open(outputfile, 'w') as file_output:
        writer = csv.writer(file_output, delimiter=';')
        header = ['ts_name', 'time', 'unit', 'value', 'class']
        writer.writerow(header)
        
def append_to_file(df, outputfile):
    with open(outputfile, 'a') as file_output:
        writer = csv.writer(file_output, delimiter=';')
        for index, row in df.iterrows():
            time_index = datetime.datetime.strptime(row[1], '%Y-%m-%d %H:%M:%S')
            time_index = time_index.replace(second=0, microsecond=0)
            row = [row[0], time_index.isoformat(), row[2], row[3], row[4]]
            writer.writerow(row)
```

```python
init_file(test_file_name)
init_file(train_file_name)
```

```python
for file in files:
    df = load_data_frame_with_labels(file['name'], file['file_name'])
    df_test = df.iloc[0:1000]
    df_train = df.iloc[1001:1626]
    append_to_file(df_test, test_file_name)
    append_to_file(df_train, train_file_name)
```
